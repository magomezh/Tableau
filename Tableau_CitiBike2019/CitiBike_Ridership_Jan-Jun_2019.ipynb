```python
import pandas as pd
from glob import glob
import datetime 
import numpy as np
```

```python
riders_data = sorted(glob("ridership/CitibikeData2019_******.csv"))
riders_data
```

```python
riders_data_2019 = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in riders_data), ignore_index = True)
riders_data_2019
```

```python
riders_data_2019 = riders_data_2019.drop(columns=['Column 1', 'filename'])
riders_data_2019.head()
```

```python
riders_data_2019.rename(columns={'24-Hour Passes Purchased (midnight to 11:59 pm)': '24-Hour Passes', '3-Day Passes Purchased (midnight to 11:59 pm)': '3-Day Passes', 'Miles traveled today (midnight to 11:59 pm)': 'Miles Traveled Today'}, inplace=True)
```

```python
riders_data_2019.head()
```

```python
riders_data_2019.rename(columns={ 'Total Annual Members (All Time)': 'Total Annual Members', 'Trips over the past 24-hours (midnight to 11:59pm)': 'Trips over the past 24-hours', }, inplace=True)
```

```python
riders_data_2019.head()
```

```python
riders_data_2019.to_csv('ridership/CitiBike_Ridership_2019.csv', sep = ',', header = True)
```
